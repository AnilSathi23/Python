## Line Plot
 A Line plot is a basic type of chart that displays information as series of points onnected by straightline segments, typically useful or showing trends over a interval

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for line plot

x =  np.linspace(0,10, 100)
y  =  np.sin(x)

# create a line plot

plt.plot(x,y, label =  'Sinewave', color =  'aqua', linestyle = 'dashed', linewidth = 2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Chart')
plt.legend(loc = 'best')
plt.savefig('test.jpg')
plt.show()
```

```python
# create a line plot
plt.plot(x,np.cos(x), label =  'Cosine Wave', color =  'green', linestyle = 'dotted', linewidth = 5, gapcolor = 'red')
#plt.show()
plt.plot(x,y, label =  'Sinewave', color =  'aqua', linestyle = 'dashed', linewidth = 10, gapcolor = 'purple',
        alpha = 0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine and  Cosine Chart')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
# Subplotting and figsize
# Change the size of plot

plt.figure(figsize = (15, 10))
plt.plot(x,y, label =  'Sinewave', color =  'aqua', linestyle = 'dashed', linewidth = 10, gapcolor = 'purple',
        alpha = 0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine and  Cosine Chart')
plt.legend(loc = 'best')
plt.grid()
plt.savefig('test2.jpg')
plt.show()
```

```python
# create a line plot
## subplotting

plt.figure(figsize = (10, 6))
plt.subplot(2,2,1)
plt.plot(x,np.cos(x), label =  'Cosine Wave', color =  'green', linestyle = 'dotted', linewidth = 5, gapcolor = 'red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('C0Sine Chart')
plt.legend(loc = 'best')

plt.subplot(2,2,4)
plt.plot(x,y, label =  'Sinewave', color =  'aqua', linestyle = 'dashed', linewidth = 10, gapcolor = 'purple',
        alpha = 0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Chart')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

## __Scatter Plot:__
A scatter plot is used to display the relationship between two continuous variables.
- Each point on the plot represents an observation, and the position of the point is determined by the values of the two variables.
- It is helpful for identifying patterns, clusters, or trends in the data.




```python
# Generate data 

x =  np.random.randn(100)
y = 2*x + np.random.randn(100)
```

```python
plt.scatter(x,y, color = 'green', marker = 'o', label = 'random data', s = 50)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Data  Scatter Chart')
plt.legend(loc = 'best')

plt.show()
```

## __Bar Chart:__
A bar chart is a common way to represent categorical data.
- It uses rectangular bars of lengths proportional to the values they represent.
- This chart is useful for comparing the sizes or frequencies of different categories.

```python
categories = ['CatA', 'CatB', 'CatC']
values = [25, 40,15]
plt.bar(categories,values, label =  'BarGraph', color = 'orange', edgecolor = 'black') 
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

## __Box Plot:__
A box plot (box-and-whisker plot) provides a graphical summary of the distribution of a dataset.
- It displays the minimum, first quartile, median, third quartile, and maximum values.
- This plot is useful for identifying outliers and understanding the spread and central tendency of the data.

```python
# Generate Data
np.random.seed(42)
data =  [np.random.normal(0,1, 100),np.random.normal(0,1.5, 100), np.random.normal(0,2, 100)]

# create boxplot

plt.boxplot(data, tick_labels = ['Grp1', 'Grp2','Grp3'], patch_artist = True, notch = True)
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Boxplot')

plt.show()
```

## __Pie Chart:__
A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions.
- Each slice represents a proportionate part of the whole, and the entire circle represents 100%.
- These charts are commonly used to visualize the distribution of a categorical variable or the relative sizes of different components within a whole.

```python
categories = ['CatA', 'CatB', 'CatC']
values = [25, 40,15]
plt.pie(values, labels =  categories, startangle = 45, autopct = '%1.1f%%', explode = [0.0, 0.0, 0.5] )

plt.title('Pie Chart')

plt.show()
```
